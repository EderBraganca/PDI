## Carregando uma imagem

<p>
<b>skimage.io.imread</b>(fname, as_gray=False, plugin=None, **plugin_args): Load an image from file.

Parameters:
<ul>
    <li>fname: str or pathlib.Path. Image file name, e.g. test.jpg or URL.</li>
    <li>as_gray: bool, optional. If True, convert color images to gray-scale (64-bit floats). Images that are already in gray-scale format are not converted.</li>
    <li>plugin: str, optional. Name of plugin to use. By default, the different plugins are tried (starting with imageio) until a suitable candidate is found. If not given and fname is a tiff file, the tifffile plugin will be used.</li>
</ul>

Returns:
<ul>
    <li>img_array: ndarray. The different color bands/channels are stored in the third dimension, such that a gray-image is MxN, an RGB-image MxNx3 and an RGBA-image MxNx4.</li>
</ul>
</p>

## Carregando uma imagem colorida

```python
cat = data.chelsea()
altura, largura, channel = cat.shape
print(f'Tipo: {type(cat)}')
print(f'Tipo dos dados: {cat.dtype}')
print(f'Altura: {altura} Largura: {largura} Canais: {channel}')
print(f'Número de dimensões: {cat.ndim}')

plt.imshow(cat)
```

```python
def colorida_cinza(img):
    nimg = img[:,:,0]*0.299 + img[:,:,1]*0.587 + img[:,:,2]*0.114
    return nimg
```

```python
nova = colorida_cinza(cat)
print(f'Dimensao cat: {cat.ndim}')
print(f'Dimensao nova: {nova.ndim}')
plt.imshow(nova, cmap='gray')
```

## Mostrando por separado cada plano da imagem

```python
f, ((ax0, ax1), (ax2, ax3)) = plt.subplots(2, 2, figsize=(10,10))
ax0.imshow(cat)
ax0.set_title("Cat")
ax0.axis('off')

ax1.imshow(cat[:,:,0], cmap='gray')
ax1.set_title("Vermelho")
ax1.axis('off')

ax2.imshow(cat[:,:,1], cmap='gray')
ax2.set_title("Verde")
ax2.axis('off')

ax3.imshow(cat[:,:,2], cmap='gray')
ax3.set_title("Azul")
ax3.axis('off')
```

## Mostrando a plano vermelho na cor vermelha

```python
cat2 = cat.copy()
cat2[:,:,1] = 0
cat2[:,:,2] = 0
plt.imshow(cat2)
```

## Escrita de imagens

```python
io.imsave('gato.png', cat2)
```

## Valores aceitos para imagens
0 - 255 onde 0 é preto e 255 é branco

0 - 1   onde 0 é preto e 1 é branco

```python
import numpy as np
```

```python
linear0 = np.linspace(0, 1, 2500).reshape((50,50))
linear1 = np.linspace(0, 255, 2500).reshape((50,50)).astype(np.uint8)

print("Linear0: ", linear0.dtype, linear0.min(), linear0.max())
print("Linear1: ", linear1.dtype, linear1.min(), linear1.max())

fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(10,10))
ax0.imshow(linear0, cmap='gray')
ax1.imshow(linear1, cmap='gray')
```
